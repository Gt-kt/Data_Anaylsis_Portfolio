```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = 'C:/Users/c/Downloads/Bank_analysis/Customer_churn_Analysis/data/Telco-Customer-Churn.csv'
data = pd.read_csv(file_path)

data.head()

# 1. Churn rate by demographics (gender and SeniorCitizen)
plt.figure(figsize = (10,6))

plt.subplot(1, 2, 1)
sns.countplot(data=data, x='gender', hue='Churn')
plt.title('Churn Rate by Gender')

plt.subplot(1,2,2)
sns.countplot(data = data, x='SeniorCitizen', hue='Churn')
plt.title('Churn Rate by Senior Citizen Status')
```

Demographic Impact on Churn:

Gender: There is no significant difference in churn rates between genders.
Senior Citizens: Senior citizens show a higher churn rate, indicating that older customers may have specific needs or challenges that affect their retention.

```python
data['tenure_group'] = pd.cut(data['tenure'], bins=[0,12,24,48,60,72],
                              labels=['0-12', '12-24', '24-48', '48-60', '60-72'])

plt.figure(figsize=(8,5))
sns.countplot(data=data, x='tenure_group', hue='Churn')
plt.title('Churn Rate byh Tenure Group')
plt.xlabel('Tenure Group (Months)')
plt.show()
```

Tenure and Churn:

Customers with a shorter tenure (0-12 months) tend to have a much higher churn rate compared to those with longer tenures, suggesting that customer retention strategies are particularly important in the early months.

```python
# 3. Churn by Contract Type
plt.figure(figsize=(8, 5))
sns.countplot(data=data, x='Contract', hue='Churn')
plt.title('Churn Rate by Contract Type')
plt.show()
```

Contract Type:

Month-to-month contracts have a notably higher churn rate compared to one-year or two-year contracts, highlighting the value of long-term commitments in reducing churn.

```python
# 4. Churn by Monthly Charges
plt.figure(figsize=(8, 5))
sns.histplot(data[data['Churn'] == 'No']['MonthlyCharges'], kde=True, label='No Churn')
sns.histplot(data[data['Churn'] == 'Yes']['MonthlyCharges'], kde=True, label='Churn', color='orange')
plt.title('Monthly Charges Distribution by Churn Status')
plt.xlabel('Monthly Charges')
plt.legend()
plt.show()
```

Monthly Charges and Churn:

Churned customers generally have higher monthly charges. This suggests that cost-sensitive customers may be more likely to leave, potentially due to perceived lack of value or affordability issues.

```python
# 5. Payment Method Impact on Churn
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='PaymentMethod', hue='Churn')
plt.title('Churn Rate by Payment Method')
plt.xticks(rotation=45)
plt.show()
```

Payment Method:

Customers using electronic checks as a payment method show a higher churn rate, while those using bank transfer or credit cards tend to have lower churn rates. This could be an area for exploration to improve retention by encouraging more stable payment methods.

Conclusion of Analysis:
The analysis reveals key patterns in customer churn behavior, offering actionable insights for retention strategies:

Customer Demographics: Senior citizens show higher churn rates, suggesting a need for tailored services or support for older customers. Gender does not play a significant role in churn.

Early Tenure Vulnerability: Churn is highest among new customers, particularly in the first 12 months, highlighting the importance of engaging customers early with onboarding and retention initiatives.

Contract Type and Retention: Month-to-month contracts are linked to higher churn rates, while longer contracts (one or two years) reduce churn. This suggests that encouraging longer commitments can improve retention.

Cost Sensitivity: Customers with higher monthly charges are more likely to churn, implying that high costs may deter customers. Providing more flexible or tailored pricing options could be a potential retention strategy.

Payment Method Influence: Electronic check users tend to have higher churn rates compared to those using automatic bank transfers or credit cards, indicating that stabilizing payment methods might correlate with better retention.
